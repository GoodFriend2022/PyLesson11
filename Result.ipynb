```python
from matplotlib import pyplot as plt
import math
```

```python
def f(x):
    y = -12 * x ** 4 * math.sin(math.cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
    return y

def diap(end):
    x = [x/10000 for x in range(-end, end + 1)]
    return x
```

> **Строим график:**

```python
x = diap(100000)
y = [f(i) for i in x]
plt.plot(x, y)
```

> **Вычисляем корни:**

```python
item = y[0]
rad = []
y_rad = []
for i in x:
    fun = f(i)
    if fun < 0:
        if item > 0:
            rad.append(i)
            y_rad.append(fun)
    elif fun > 0:
        if item < 0:
            rad.append(i)
            y_rad.append(fun)
    item = fun
print(f'Примерные корни в заданном диапазоне по оси "х": {rad}')    
plt.scatter(rad, y_rad)
```

> **Находим промежутки на которых функция либо больше либо меньше нуля:**

```python
poz = []
neg = []
if y[0] < 0:
    count = 1
    neg.append(f'до {round(rad[0], 2)}')
else: 
    count = 0
    poz.append(f'до {round(rad[0], 2)}')
for i in range(len(rad) - 1):
    if count == 1:
        poz.append(f'от {round(rad[i], 2)} до {round(rad[i+1], 2)}')
        count = 0
    else:
        neg.append(f'от {round(rad[i], 2)} до {round(rad[i+1], 2)}')
        count = 1
if y[0] < 0:
    count = 1
    neg.append(f'от {round(rad[len(rad) - 1], 2)}')
else: 
    count = 0
    poz.append(f'от {round(rad[len(rad) - 1], 2)}')
print(f'f(x)>0 в интервалах по оси "х": {poz}\n')
print(f'f(x)<0 в интервалах по оси "х": {neg}')
```

> **Определяем вершины:**

```python
def orient():
    if y[0] > y[1]:
        return 'neg'
    else: return 'poz'
item = orient()
top_y = []
top_x = []
for i in range(2, len(y)-1):
    if item == 'neg':
        if y[i] > y[i - 1]:
            top_x.append(round(x[i], 2))
            top_y.append(round(y[i], 2))
            item = 'poz'
    else:
        if y[i] < y[i - 1]:
            top_x.append(round(x[i], 2))
            top_y.append(round(y[i], 2))
            item = 'neg'
coord = [(top_x[i], top_y[i]) for i in range(len(top_x))]
print(f'Примерные координаты вершин: {coord}')
plt.scatter(top_x, top_y)
```

> **Находим интервалы, на которых функция либо возрастает либо убывает:**

```python
item = orient()
int_poz = []
int_neg = []
if item == 'poz':
    int_poz.append(f'до {top_x[0]}')
else: int_neg.append(f'до {top_x[0]}')
for i in range(len(top_x) - 1):
    if item == 'poz':
        int_neg.append(f'от {top_x[i]} до {top_x[i + 1]}')
        item = 'neg'
    else:
        int_poz.append(f'от {top_x[i]} до {top_x[i + 1]}')
        item = 'poz'
if item == 'poz':
    int_neg.append(f'от {top_x[len(top_x) - 1]}')
else: int_poz.append(f'от {top_x[len(top_x) - 1]}')
print(f'Интервалы, на которых функция возрастает: {int_poz}\n')
print(f'Интервалы, на которых функция убывает: {int_neg}')     
```
